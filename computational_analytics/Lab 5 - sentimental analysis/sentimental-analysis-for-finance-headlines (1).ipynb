```python
df = pd.read_csv('../input/sentiment-analysis-for-financial-news/all-data.csv',names = ['lables','messages'],encoding='ISO-8859-1')
df.head()
```

```python
# checking for any empty cell
df.isnull().sum()
```

```python
# Shape of our dataset
df.shape
```

```python
df_copy = df.copy()
df_copy.head()
```

```python
def func(df_new):
    if df_new == 'neutral':
        return 0
    elif df_new == 'negative':
        return 1
    else:
        return 2
df_copy['target'] = df_copy.lables.apply(func)
df_copy.head()
```

```python
# Maptplolib
import matplotlib.pyplot as plt 

# Seaborn
import seaborn as sns 
```

```python
# Count plot for labels feature
plt.figure(figsize=(9,7))
df_copy.lables.value_counts().plot(kind='bar',color=['blue','lime','red'])
plt.xlabel("News Type")
plt.ylabel("count")
plt.title("Count Plot for labels");
```

```python
#nltk
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
#Regular Expressions
import re
```

```python
# Text preprocessing
corpous = list()

#Object for porterstem
ps = PorterStemmer()

for i in range(len(df_copy)):
    text = re.sub(r'[^a-zA-Z]',' ',df_copy.messages[i])
    text = text.lower()
    text = text.split()
    text = [ps.stem(word) for word in text if word not in set(stopwords.words('english'))]
    text = ' '.join(text)
    corpous.append(text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,ngram_range=(1,3))
X = cv.fit_transform(corpous).toarray()
```

```python
# Dependent Feature
y = df_copy.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=43)
```

```python
cv.get_feature_names()[:10]
```

```python
# Scikit-learn
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
# Build Models

models = {
    MultinomialNB():'Multinomial Naive Bayes',
    LogisticRegression(max_iter=300):'Logistic Regression',
    SVC(C=1.0, kernel='linear', degree=3, gamma='auto'):"Support Vector Machine"
}
for m in models.keys():
    m.fit(X_train,y_train)
```

```python
# Check Accuracy for each model
for model,name in models.items():
     print(f"Accuracy Score for {name} is : ",model.score(X_test,y_test)*100,"%")
```
