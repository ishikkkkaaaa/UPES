```python
from nltk.corpus import indian
```

```python
final_list = indian.sents()
```

```python
final_list
```

```python
flat = []
for i in final_list:
    for j in i:
        flat.append(j)
```

```python
flat
```

```python
from nltk.util import ngrams
```

```python
n=1
unigrams = ngrams(flat,n)
```

```python
unigrams
```

```python
for i in unigrams:
    print(i)
```
